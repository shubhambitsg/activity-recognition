```python
all_activity_scores = {}
def concatScores(scoreDir):
    files = os.listdir(scoreDir)
    for score_json_files in files:
        fname = os.path.join(os.path.abspath(DIR),score_json_files)
        with open(fname) as data_file:
            data = json.load(data_file)
            for act,scores in data.items():
                if act in list(all_activity_scores.keys()):
                    score,fname = calculate_max_prob(act,scores,score_json_files,all_activity_scores)
                    all_activity_scores[act] = [score,fname]
                else:
                    all_activity_scores[act] = [scores,score_json_files]
    return all_activity_scores
    
        
def calculate_max_prob(act,scores,score_file,all_activities_scores):
        if  all_activities_scores[act] and all_activities_scores[act][0] < scores:
            return scores,score_file
        else:
            return all_activities_scores[act][0],all_activities_scores[act][1]
        

    
```

```python
import spacy
token_tags = ['VBG', 'VBZ', 'VBD', 'VB']
def verbParsing(doc):
    return([(token.text) for token in doc if token.tag_ in token_tags])
```

```python
try:
    nlp
except:
    nlp = spacy.load('en_core_web_md')
```

```python
# sentence = 'Jack loves climbing and bicycling on Sundays.'
# sentence = 'I like to surf and exercise regularly'
sentence = 'Michael Jordan was the best when it comes to dunking'


doc = nlp(sentence)
actionList = verbParsing(doc)
actionList
```

```python
def get_related(word):
  filtered_words = [w for w in word.vocab if w.is_lower == word.is_lower and w.prob >= -15]
  similarity = sorted(filtered_words, key=lambda w: word.similarity(w), reverse=True)
  return similarity[:20]
```

```python
def tag_synonyms(action):
    synList=  [w.lower_ for w in get_related(nlp.vocab[action])]
    return synList
```

```python
predictions = concatScores(DIR)
predicted_actions = set(concatScores(DIR).keys())
```

```python
tagged_actions = {action: set(tag_synonyms(action)) for action in actionList}
```

```python
vid_name = []
for action in actionList:
    syns = tagged_actions[action]
    found_actions = predicted_actions.intersection(syns)
    found_predictions = [tuple(predictions[act]) for act in list(found_actions)]
    found_predictions = sorted(found_predictions, key=lambda kv: kv[0], reverse=True)
#     for prob, scene in found:
#         print(prob, scene)
    if len(found_predictions) > 0:
        score, scene = found_predictions[0]
        print(action, scene, score)
        vid_name.append((action,scene))
```

```python
with open('/tmp/concat.txt', 'w') as f:
    for activity, vid in vid_name:
        fl = (os.path.abspath(DIR+'../scene/'+ vid.split('.')[0] + '.mp4'))
        f.write('file '+"'" +fl + "'")
        f.write('\n')
```
